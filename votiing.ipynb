```python
def get_max_freq_value(row):
    counts = row[['catboost', 'xgb', 'lgbm']].value_counts()
    if counts.get(0, 0) == counts.get(1, 0):  # If there is a tie
        prob_0 = np.mean([row['y_prob_catboost'], row['y_prob_xgb'], row['y_prob_lgbm']] * (row[['catboost', 'xgb', 'lgbm']] == 0))
        prob_1 = np.mean([row['y_prob_catboost'], row['y_prob_xgb'], row['y_prob_lgbm']] * (row[['catboost', 'xgb', 'lgbm']] == 1))
        return 0 if prob_0 > prob_1 else 1
    return counts.idxmax()

# Applying the function to each row
df['max_freq_value'] = df.apply(get_max_freq_value, axis=1)
df['max_freq_value'] = df['max_freq_value'].astype(int)
```

```python
y_pred = np.array(df["max_freq_value"])
```

```python
y_pred = np.array(df["max_freq_value"])
y_test = np.array(df["y_actual"])

correct_predictions = np.sum(y_pred == y_test)

# Calculate the total number of predictions
total_predictions = len(y_test)

# Calculate the accuracy
accuracy = correct_predictions / total_predictions

print(f'Manual Accuracy: {accuracy:.4f}')
```
